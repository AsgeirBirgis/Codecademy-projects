```python
import pandas as pd
from matplotlib import pyplot as plt
csv_file_path = "C:/Users/asgei/Downloads/Life-Expectancy-and-GDP-Starter/Life-Expectancy-and-GDP-Starter/all_data.csv"
df = pd.read_csv(csv_file_path)
print(df.head())
```

```python
unique_countries = df['Country'].unique()

print(unique_countries)
```

```python
#Let's see what the average life expectancy is by country and the average GDP.
```

```python
average_life_expectancy_by_country = df.groupby('Country')['Life expectancy at birth (years)'].mean()


print(average_life_expectancy_by_country)
```

```python
average_gdp_by_country = df.groupby('Country')['GDP'].mean()

print(average_gdp_by_country)
```

```python
#Let's see how GDP and Life expectancy have evolved in the countries over the years the data span. 
```

```python
grouped_data = df.groupby('Country')

for country, data in grouped_data:
    plt.plot(data['Year'], data['Life expectancy at birth (years)'], label=country)

plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy Evolution in Seven Countries')

plt.legend()
plt.show()
plt.close()
```

```python
#As we can see, life expectancy differs between counties, with it being highest in Germany and lowest in Zimbabwe. 
#Life expectancy is increasing in all countries, but this trend is most prominently in Zimbabwe. Now let's do the same for GDP. 
```

```python
grouped_data = df.groupby('Country')

for country, data in grouped_data:
    plt.plot(data['Year'], data['GDP'], label=country)

plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('Evolution of GDP in Seven Countries')

plt.legend()
plt.show()
plt.close()
```

```python
#As we can see, GDP has risen significantly in the United States and China. Other countries have remained more stable. 
#Interestingly enough, although Zimbabwe has seen a positive trend in life expectancy, that trend is not mirrored in GDP in this
#graph. That might be because how large the numbers are for the other countries might be obscuring the change that has happened 
#in this period in Zimbabwe. Let's zoom in on Zimbabwe and see the trend.
```

```python
zimbabwe_data = df[df['Country'] == 'Zimbabwe']

plt.plot(zimbabwe_data['Year'], zimbabwe_data['GDP'], marker='o', linestyle='-', label='Zimbabwe')

plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP Evolution in Zimbabwe')

plt.legend()

plt.show()
```

```python
zimbabwe_data = df[df['Country'] == 'Zimbabwe']

min_gdp = zimbabwe_data['GDP'].min()
max_gdp = zimbabwe_data['GDP'].max()

print("Lowest GDP in Zimbabwe:", min_gdp)
print("Highest GDP in Zimbabwe:", max_gdp)
```

```python
#As we can see, the GDP in Zimbabwe has changed significantly over this time, with the lowest coming in 2008 at 4.415.702.800$ 
#and the highest coming in 2016 at 16.304.667.807$. So though these are not big numbers compared to the world's industrial 
#leaders, Zimbabwe has almost quadrupled its GDP from 2008 to 2016. Interestingly the trend in GDP is in part mirrored by 
#the direction in life expectancy. Life expectancy started increasing in 2005 up until 2016. At the same time, GDP takes a big 
#step upward from 2008 til 2016. 
```
